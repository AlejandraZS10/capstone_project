```python
import pandas as pd
import geopandas as gpd
#truck routes
ct = gpd.read_file('data/2010 Census Tracts.geojson')
ct
```

```python
ct= ct[['boro_name','ctlabel','geometry']]
```

```python
ss = gpd.read_file('sel_streets.geojson')
ss.head(3)
```

```python
streets_with_ct = gpd.sjoin(ss, ct, how="left", op='intersects')
streets_with_ct.head()
```

```python
streets_with_ct.plot()
```

```python
streets_with_ct = pd.DataFrame(streets_with_ct)
streets_with_ct
```

```python
streets_with_ct['unique_id'] = range(len(streets_with_ct))
streets_with_ct['unique_id']
```

```python
t=streets_with_ct[['unique_id','ctlabel','total']]
```

```python
results = []
grouped = t.groupby(['ctlabel'])
for key, group in grouped:
    for idx, row in group.iterrows():
        # The group excluding current row
        group_other = group.drop(idx)  
        sum_tots = group_other['total'].sum()
        results.append(row.tolist() + [sum_tots])
        
results_df = pd.DataFrame(
    results, columns=['unique_id','ctlabel', 'total', 'sum_total']
)
results_df
```

```python
test = streets_with_ct.merge(results_df, how='left', on=['unique_id','ctlabel','total'])
test
```

```python
geo = gpd.GeoDataFrame(test, geometry='geometry')
geo.to_file("data.geojson", driver='GeoJSON')
```
