A geojson file with all streets in the five boroughs of New York. The file serves as a database of all existing streets.

This file adds a new unique ID to every street to keep track of spatial connections with other databases.

```python
import pandas as pd
import geopandas as gpd
#all streets in New York
scl = gpd.read_file('data/NYC Street Centerline (CSCL).geojson')
scl
```

```python
scl.plot()
```

```python
scl.info()
```

```python
scl= scl[['rw_type','bike_lane','snow_pri','borocode','st_label','physicalid','st_width','rw_type','geometry']]
scl
```

```python
#rename columns
scl = scl.rename(columns={
    'rw_type':'roadway_type',
    'snow_pri':'snow_priority',
    'borocode':'boro_code',
    'st_label':'roadway_name',
    'physicalid':'street_id'
})
```

```python
#fix nones
scl['bike_lane'] = scl['bike_lane'].fillna(0)
scl['snow_priority'] = scl['snow_priority'].fillna('N')
```

```python
#lower case
scl['roadway_name'] = scl['roadway_name'].str.lower()
```

```python
#fix abbreviations
scl['roadway_name'] = scl['roadway_name'].str.replace(' blvd', ' boulevard') 
scl['roadway_name'] = scl['roadway_name'].str.replace(' e ', 'east ') 
scl['roadway_name'] = scl['roadway_name'].str.replace(' w ', 'west ') 
scl['roadway_name'] = scl['roadway_name'].str.replace(' pkwy', ' parkway') 
scl['roadway_name'] = scl['roadway_name'].str.replace(' pl', ' place') 
scl['roadway_name'] = scl['roadway_name'].str.replace(' st', ' street') 
scl['roadway_name'] = scl['roadway_name'].str.replace(' rd', ' road')
scl['roadway_name'] = scl['roadway_name'].str.replace(' ave', ' avenue')
scl['roadway_name'] = scl['roadway_name'].str.replace('ave ', 'avenue ')
scl['roadway_name'] = scl['roadway_name'].str.replace(' ln', ' lane')
scl['roadway_name'] = scl['roadway_name'].str.replace(' dvwy', ' driveway')
scl['roadway_name'] = scl['roadway_name'].str.replace(' dr', ' drive')
scl['roadway_name'] = scl['roadway_name'].str.replace(' expy', ' expresspway')
scl['roadway_name'] = scl['roadway_name'].str.replace(' tunl', ' tunnel')
scl['roadway_name'] = scl['roadway_name'].str.replace(' brg', ' bridge')
scl['roadway_name'] = scl['roadway_name'].str.replace(' sq', ' square')
scl['roadway_name'] = scl['roadway_name'].str.replace(' plz', ' plaza')
scl['roadway_name'] = scl['roadway_name'].str.replace(' cir', ' circle')
scl['roadway_name'] = scl['roadway_name'].str.replace(' tpke', ' turnpike')
scl['roadway_name'] = scl['roadway_name'].str.replace(' cres', ' crescent')
scl['roadway_name'] = scl['roadway_name'].str.replace(' fwy', ' freeway')
scl['roadway_name'] = scl['roadway_name'].str.replace(' appr ', ' approach')
scl['roadway_name'] = scl['roadway_name'].str.replace(' lk', ' lake')
scl['roadway_name'] = scl['roadway_name'].str.replace(' hwy', ' highway')
scl['roadway_name'] = scl['roadway_name'].str.replace(' hts', ' heights')
scl['roadway_name'] = scl['roadway_name'].str.replace(' grn', ' green')
scl['roadway_name'] = scl['roadway_name'].str.replace(' pth', ' path')
scl['roadway_name'] = scl['roadway_name'].str.replace(' drwy', ' driveway')
scl['roadway_name'] = scl['roadway_name'].str.replace(' rdg', ' ridge')
```

```python
scl
```

```python
geo = gpd.GeoDataFrame(scl, geometry='geometry')
geo.to_file("fixed_streets.geojson", driver='GeoJSON')
```
